```python
#Representación numérica de una variable categórica
def to_onehot(x):
    values = np.unique(x)
    r = np.r_[[np.argwhere(i==values)[0][0] for i in x]]
    return np.eye(len(values))[r].astype(int)
```

```python
def replace_columns_with_onehot(df_sin_nulos, col):
    k = to_onehot(df_sin_nulos[col].values)
    r = pd.DataFrame(k, columns=["%s_%d"%(col, i) for i in range(k.shape[1])], index=df_sin_nulos.index).join(df_sin_nulos)
    del(r[col])
    return r
```

```python
d = replace_columns_with_onehot(df_sin_nulos, "FAMI_EDUCACIONPADRE")
d = replace_columns_with_onehot(d, "FAMI_EDUCACIONMADRE")
d.head()
```

```python
##Normalizar valores para ajustarlos a una escala específica
scaler = MinMaxScaler()
d['ESTU_VALORMATRICULAUNIVERSIDAD'] = scaler.fit_transform(d[['ESTU_VALORMATRICULAUNIVERSIDAD']])
d['ESTU_HORASSEMANATRABAJA'] = scaler.fit_transform(d[['ESTU_HORASSEMANATRABAJA']])
d['FAMI_ESTRATOVIVIENDA'] = scaler.fit_transform(d[['FAMI_ESTRATOVIVIENDA']])
d.head()
```
